```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('C:/Users/DeLL/OneDrive/ML Datasets/store_data.csv', header = None)
store_data.head()
```

```python
store_data.shape
```

```python
store_data.replace(np.nan, 0, inplace = True)
store_data.head()
```

```python
records = []
for i in range(0,7501):
    x = list(set(store_data.loc[i]))
    records.append(x)
```

```python
for i in records:
    for j in i:
        if j == 0:
            i.remove(j)
```

```python
records
```

## Apriori

```python
association_rules = apriori(records, min_support = 0.0045, min_confidence = 0.2, min_lift = 4, max_length = 3)

association_results = list(association_rules)
print(len(association_results))
```

```python
print(association_results[0])
```

```python
print(association_results[0][1])
```

## Dataset_2

```python
gs_data = pd.read_csv('C:/Users/DeLL/OneDrive/ML Datasets/GroceryStoreDataset.csv', header = None)
gs_data
```

```python
records = []
for i in range(0,20):
    x = list(gs_data.loc[i])
    records.append(x)
```

```python
ls = []
for i in records:
    for j in i:
        j = j.split(',')
        ls.append(j)
ls
```

```python
association_rules = apriori(ls, min_support = 0.0045, min_confidence = 0.2, min_lift = 2, max_length = 3)

association_results = list(association_rules)
print(len(association_results))
```
